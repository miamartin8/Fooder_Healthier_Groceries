1. Combining Original Data Categories with New Clusters for Null Values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
new_cat = pd.read_csv('Null_Categories_Fixed.csv')
```

```python
nutrition = pd.read_csv('Nutrition_Data_Branded_Food.csv')
```

```python
nutrition.head()
```

```python
new_cat.head()
```

```python
new_cat.shape
```

```python
nutrition.shape
```

```python
new_cat = new_cat[['fdc_id', 'cluster']]
```

```python
new_cat.head()
```

```python
final_nutrition = pd.merge(left=nutrition, right=new_cat, how='outer', on='fdc_id')
```

```python
final_nutrition.shape
```

```python
final_nutrition.isnull().sum()
```

```python
final_nutrition.head()
```

```python
category_numbers = dict(zip([i for i in set(final_nutrition['branded_food_category'].values)],[i + 1000 for i in range(len(set(final_nutrition['branded_food_category'].values)))]))
```

```python
final_nutrition['new_cluster'] = final_nutrition[['branded_food_category', 'cluster']].apply(lambda x: x[1] if x[0]=='0' else category_numbers[x[0]], axis=1)
```

```python
final_nutrition.head()
```

```python
final_nutrition.to_csv('Final_Nutrition.csv', index=False)
```
