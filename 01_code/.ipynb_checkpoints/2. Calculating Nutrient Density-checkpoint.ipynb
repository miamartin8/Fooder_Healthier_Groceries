### Issues with this calculation

- Iron and salt seem to be really driving the calculations
- Soy beans are ranking the highest on the list due to incredibly high iron
- Candy corn is ranking the lowest on the list due to incredibly high salt (is this data right??)
- Maybe these amounts need to be capped.  There is a limit at which high amounts of vitamins are useful.  
- Also, the vitamins are outweighing important items like protein and fiber - need a way to weight these values appropriately

**TEST** Cap at 1 (100% of DV recomended) - suggested in article

```python
branded.shape
```

```python
branded.protein_grams.clip(upper=1, inplace=True)
branded.fiber_grams.clip(upper=1, inplace=True)
branded.vitamin_a_IU.clip(upper=1, inplace=True)
branded.vitamin_c_MG.clip(upper=1, inplace=True)
branded.calcium_MG.clip(upper=1, inplace=True)
branded.iron_MG.clip(upper=1, inplace=True)
branded.potassium_mg.clip(upper=1, inplace=True)
branded.magnesium_mg.clip(upper=1, inplace=True)
branded.mufa_grams.clip(upper=1, inplace=True)
branded.sugar_G.clip(upper=1, inplace=True)
branded.sat_fat_g.clip(upper=1, inplace=True)
branded.sodium_MG.clip(upper=1, inplace=True)
branded.trans_fat_g.clip(upper=1, inplace=True)
```

```python
branded['NRFNn.3'] = branded['protein_grams']+branded['fiber_grams']+branded['vitamin_a_IU']+branded['vitamin_c_MG']+branded['calcium_MG']+branded['mufa_grams']+branded['potassium_mg']+branded['magnesium_mg']+branded['iron_MG']-branded['trans_fat_g']-branded['sat_fat_g']-branded['sugar_G']-branded['sodium_MG']
```
