```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
final = pd.read_csv('Final_Nutrition.csv')
```

```python
final.set_index('fdc_id', inplace=True, drop=False)
```

```python
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

print(color.RED + 'Hello World !' + color.END)
```

```python
q = input('Enter Food Here: ')
df = pd.DataFrame([])
print(final[final['description'].str.contains(q)]['description'])
inp = int(input('Choose Food Item: '))
cluster = final[final['new_cluster']==final['new_cluster'][inp]]
print(color.RED + color.BOLD + 
      f"Your original food selection was {final['description'][inp]} with a nutrition index of {final['NRFNn.3'][inp]:.2f}." + 
      color.END)
for i in cluster.index:
    if final['NRFNn.3'][i] > final['NRFNn.3'][inp]:
        df = df.append(pd.DataFrame({'ID': final['fdc_id'][i], 
                                     'Alternative Food Option': final['description'][i], 
                                     'Nutrition Index': final['NRFNn.3'][i], 
                                     'Protein %DV': final['protein_grams'][i]*100, 
                                     'Fiber %DV': final['fiber_grams'][i]*100, 
                                     'Monounsaturated Fat %DV': final['mufa_grams'][i]*100,
                                     'Vitamin A %DV': final['vitamin_a_IU'][i]*100, 
                                     'Vitamin C %DV': final['vitamin_c_MG'][i]*100, 
                                     'Calcium %DV': final['calcium_MG'][i]*100, 
                                     'Iron %DV': final['iron_MG'][i]*100, 
                                     'Potassium %DV': final['potassium_mg'][i]*100, 
                                     'Saturated Fat %DV': final['sat_fat_g'][i]*100, 
                                     'Trans Fat %DV': final['trans_fat_g'][i]*100, 
                                     'Sugar %DV': final['sugar_G'][i]*100, 
                                     'Sodium %DV': final['sodium_MG'][i]*100},
                                    index=[0]), ignore_index=True)
    else:
        pass
if df.empty:
    print("You selected the healthiest item!")
else:
    pd.set_option('display.max_colwidth', -1)
    df = df.round(2)
    df = df.sort_values(by='Nutrition Index', ascending=False)
    print(color.RED + color.BOLD + "Would you consider one of the following healthier options?" + color.END)
df
```
