```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nutrition = pd.read_csv('Final_Nutrition.csv')
```

```python
vis = pd.read_csv('Test_Data_v10.csv')
```

```python
vis['Nutrition_Improvement'] = vis['Alternative_Nutrition'] - vis['Original_Nutrition']
```

```python
vis['Nutrition_Percent_Change'] = round((vis['Nutrition_Improvement'] / abs(vis['Original_Nutrition']))*100, 2)
```

```python
vis['Original_Nutrition'] = vis['Original_Nutrition'].map(lambda x: 0.01 if x == 0.000000 else x)
```

```python
vis['Nutrition_Percent_Change'] = round((vis['Nutrition_Improvement'] / abs(vis['Original_Nutrition']))*100, 2)
```

```python
vis.sort_values(by='Nutrition_Percent_Change', ascending=False).head()
```

```python
vis.describe()
```

```python
plt.figure(figsize=(15,10))
plt.hist(vis['Original_Nutrition'], color='orange', alpha=0.7, label="Original Item Nutrient Density")
plt.hist(vis['Alternative_Nutrition'], color='purple', alpha=0.5, label="Recommended Item Nutrient Density")
plt.title("Frequency of Nutrient Density Before and After Recommender", size=19)
plt.xlabel("Nutrient Density Score", size=15)
plt.ylabel("Frequency", size=15)
plt.legend(prop={'size': 15})
plt.savefig('Frequency_Plot.png')
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x = vis['Original_Nutrition'], y = vis['Alternative_Nutrition'], color='purple')
plt.title("Recommended Item Nutrient Density vs. Original Item Nutritient Density", size=19)
plt.xlabel("Original Item Nutrient Density Score", size=15)
plt.ylabel("Recommended Item Nutrient Density Score", size=15)
plt.grid(alpha=0.3)
plt.savefig('Scatter_Plot.png')
```
